## Ames Housing Sale Price Prediction - Feature Engineering

We will be performing all the below steps in Feature Engineering

1. Missing values
2. Temporal variables
3. Categorical variables: remove rare labels
4. Standarise the values of the variables to the same range

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

pd.pandas.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('Ames_train.csv')
data.shape
```

```python
#Splitting into train and test to avoid data leakage

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, data['SalePrice'], test_size=0.1, random_state=0)
X_train.shape, X_test.shape
```

## Fixing Missing Values

```python
#Lets see the categorical features first
cat_features_nan = [feature for feature in data.columns 
                             if data[feature].dtypes == 'O' and data[feature].isnull().sum() > 0]

for feature in cat_features_nan:
    print('%s: %.2f%% null values' %(feature, data[feature].isnull().sum() * 100/len(data)))
```

```python
data['Alley'].isnull().sum()
```

```python
#replace nan with the text 'Missing'

data[cat_features_nan] = data[cat_features_nan].fillna('Missing')
data.head()
```

```python
data.isnull().sum()
```

```python
data[cat_features_nan].isnull().sum()
```

```python
#Lets check the numericals that have nan
nums_features_nan = [feature for feature in data.columns 
                              if data[feature].dtypes != 'O' and data[feature].isnull().sum() > 0]

for feature in nums_features_nan:
    print('%s feature has %.3f%% null values' %(feature, data[feature].isnull().sum() * 100 / len(data)))
```

```python
#lets replace the numerical missing values with median since we have outliers

for feature in nums_features_nan:
    median_value = data[feature].median()
    data[feature + '_nan'] = np.where(data[feature].isnull(), 1, 0) #has to be first
    data[feature] = data[feature].fillna(median_value)    
```

```python
data.head()
```

```python
data[nums_features_nan].isnull().sum()
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:       
    data[feature]=data['YrSold']-data[feature]
```

```python
data.head()
```

## Continuous Numerical Variables - removing skewness

```python
# We shall remove skewness only for continuous numerical variables 
# only the 5 features below don't have a 0 among its values
num_features_cont =['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features_cont:
    data[feature] = np.log(data[feature])
    
data.head()
```

## Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
cat_features = [feature for feature in data.columns if data[feature].dtypes == 'O']
cat_features
```

```python
data['SaleCondition'].value_counts()
```

```python
#Make the category with less than 1% of values to be rare

cat_features1 = cat_features[:2]
for feature in cat_features1:
    temp=data.groupby(feature)['SalePrice'].count()/len(data)
    print(temp)
    
```

```python
#Make the category with less than 1% of values to be rare

cat_features1 = cat_features[:2]
for feature in cat_features1:
    temp=data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df = temp[temp>0.01].index
    print(temp_df)
    
```

```python
#Make the category with less than 1% of values to be rare

for feature in cat_features:
    temp=data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df = temp[temp>0.01].index
    data[feature] = np.where(data[feature].isin(temp_df), data[feature], 'Rare_Value')
    
```

```python
data.head(100)
```

```python
# Each categorical value is sorted in the ascending order wrt to mean sale price and then it is mapped to its rank
# This is called mean encoding

cat_features1 = cat_features[:5]
for feature in cat_features1:
    labels_ordered = data.groupby(feature)['SalePrice'].mean().sort_values().index
    print(labels_ordered)
```

```python
# Each categorical value is sorted in the ascending order wrt to mean sale price and then it is mapped to its rank
# This is called mean encoding

cat_features1 = cat_features[:3]
for feature in cat_features1:
    labels_ordered = data.groupby(feature)['SalePrice'].mean().sort_values().index
    print(labels_ordered)
    labels_ordered = {k:i for i, k in enumerate(labels_ordered, 0)}
    print(labels_ordered)
    print('--' * 40)
```

```python
# Each categorical value is sorted in the ascending order wrt to mean sale price and then it is mapped to its rank
# This is called mean encoding (This removes missing values also as they are numbered to some value)

for feature in cat_features:
    labels_ordered = data.groupby(feature)['SalePrice'].mean().sort_values().index    
    labels_ordered = {k:i for i, k in enumerate(labels_ordered, 0)}    
    data[feature] = data[feature].map(labels_ordered)

data.head()
```

## Feature Scaling

At this point the categorical features are also marked as Numericals, hence scaling them 

```python
features_2_scale=[feature for feature in data.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[features_2_scale])
```

```python
scaler.transform(data[features_2_scale])
```

```python
data.head()
```

```python
data1 = pd.DataFrame(scaler.transform(data[features_2_scale]), columns = features_2_scale)
data1.head()
```

```python
data = pd.concat([data[['Id', 'SalePrice']], data1], axis=1)
data.head()
```
